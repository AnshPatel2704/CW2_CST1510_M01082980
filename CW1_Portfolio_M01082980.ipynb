```python
# -------------------------------- PART 1: Create an empty list -------------------------------- #
# This list will store all positive numbers entered by the user
numbers = []

# -------------------------------- PART 2: User Input Loop -------------------------------- #
# Loop continues until the user types "done"
while True:
    user_input = input("Enter a positive number (or type 'done' to stop): ").lower()
    # If the user finishes input
    if user_input == "done":
        break
    # Check if input is a digit
    if user_input.isdigit():
        num = int(user_input)
        # Only allow positive numbers
        if num > 0:
            numbers.append(num)
        else:
            print("Please enter a positive number (greater than 0).")
    else:
        print("Invalid input. Please enter a positive number or 'done'.")

# -------------------------------- PART 3: Display Results -------------------------------- #
print("\n--- Results ---")
# Check if any valid numbers were entered
if numbers:
    print("Numbers entered:", numbers)
    # Initialize variable to store first even number
    even_number = None
    # Find the first even number
    for n in numbers:
        if n % 2 == 0:
            even_number = n
            break
    # Display result
    if even_number:
        print("First even number:", even_number)
    else:
        print("No even numbers found.")
# If no numbers were added
else:
    print("No numbers were entered.")
```

**Explanation**

• The program starts by creating an empty list called **numbers**, which is used to store all the positive numbers entered by the user.

• A **while True loop** is used to repeatedly ask the user for input.  
  This loop continues running until the user types **"done"**.

• The **input()** function is used to take user input.  
  The input is converted to lowercase so it can correctly compare with the word **"done"**.

• If the user types **"done"**, the program stops taking inputs using **break**.

• The program checks if the input contains only digits using **isdigit()**.  
  If it does, the value is converted to an integer using **int()**.

• Only **positive numbers** (greater than 0) are accepted.  
  If valid, the number is added to the **numbers list** using **append()**.  
  Otherwise, an appropriate message is shown.

• If the input is neither a positive number nor "done", the program prints an **error message**.

• After exiting the loop, the program displays **"Results"**.

• If the list is not empty, it prints all the stored numbers.  
  Then it tries to find the **first even number** by looping through the list.  
  The first number that satisfies **n % 2 == 0** is saved and displayed.

• If no even number exists, it prints a message saying so.

• If the user did not enter any numbers, it prints **"No numbers were entered."**

```python
# ------------------------------ PART 1: FUNCTION DEFINITION ------------------------------ #
# Define the function that will handle all scientific unit conversions
def measurements():

    # ------------------------------ PART 2: MAIN MENU ------------------------------ #
    # Display the main menu of conversion categories
    print("Scientific Unit Converter")
    print("1. Length")
    print("2. Temperature")
    print("3. Mass")
    print("4. Volume")
    print("5. Pressure")
    print("6. Energy")

    # Ask the user to select a category
    option = input("Enter your choice (1-6): ")

    # ------------------------------ PART 3: LENGTH CONVERSIONS ------------------------------ #
    if option == "1":
        # Show length conversion options
        print("1. Meters to Feet\n2. Inches to Centimeters\n3. Miles to Kilometers")
        choice = input("Choose a conversion (1-3): ")

        # Store conversion formulas using lambda functions
        conversions = {
            "1": lambda x: x * 3.28084,  # meters → feet
            "2": lambda x: x * 2.54,     # inches → centimeters
            "3": lambda x: x * 1.60934   # miles → kilometers
        }

        # Take numerical input from the user
        val = float(input("Enter value: "))

        # Perform the conversion if the choice is valid
        if choice in conversions:
            print("Result =", conversions[choice](val))
        else:
            print("Invalid choice!")

    # ------------------------------ PART 4: TEMPERATURE CONVERSIONS ------------------------------ #
    elif option == "2":
        print("1. Celsius to Fahrenheit\n2. Celsius to Kelvin")
        choice = input("Choose a conversion (1-2): ")
        conversions = {
            "1": lambda x: (x * 9/5) + 32,  # Celsius → Fahrenheit
            "2": lambda x: x + 273.15       # Celsius → Kelvin
        }
        val = float(input("Enter Celsius value: "))
        if choice in conversions:
            print("Result =", conversions[choice](val))
        else:
            print("Invalid choice!")

    # ------------------------------ PART 5: MASS CONVERSIONS ------------------------------ #
    elif option == "3":
        print("1. Kilograms to Pounds\n2. Grams to Ounces")
        choice = input("Choose a conversion (1-2): ")
        conversions = {
            "1": lambda x: x * 2.20462,  # kg → lbs
            "2": lambda x: x / 28.3495   # g → oz
        }
        val = float(input("Enter value: "))
        if choice in conversions:
            print("Result =", conversions[choice](val))
        else:
            print("Invalid choice!")

    # ------------------------------ PART 6: VOLUME CONVERSIONS ------------------------------ #
    elif option == "4":
        print("1. Liters to Gallons\n2. Milliliters to Ounces")
        choice = input("Choose a conversion (1-2): ")
        conversions = {
            "1": lambda x: x * 0.264172,  # liters → gallons
            "2": lambda x: x / 29.5735    # milliliters → ounces
        }
        val = float(input("Enter value: "))
        if choice in conversions:
            print("Result =", conversions[choice](val))
        else:
            print("Invalid choice!")

    # ------------------------------ PART 7: PRESSURE CONVERSIONS ------------------------------ #
    elif option == "5":
        print("1. Pascal to Bar\n2. PSI to kPa")
        choice = input("Choose a conversion (1-2): ")
        conversions = {
            "1": lambda x: x / 100000,    # Pascal → Bar
            "2": lambda x: x * 6.89476    # PSI → kPa
        }
        val = float(input("Enter value: "))
        if choice in conversions:
            print("Result =", conversions[choice](val))
        else:
            print("Invalid choice!")

    # ------------------------------ PART 8: ENERGY CONVERSIONS ------------------------------ #
    elif option == "6":
        print("1. Joules to Calories\n2. Kilojoules to Kilocalories")
        choice = input("Choose a conversion (1-2): ")
        conversions = {
            "1": lambda x: x / 4.184,  # Joules → Calories
            "2": lambda x: x / 4.184   # Kilojoules → Kilocalories
        }
        val = float(input("Enter value: "))
        if choice in conversions:
            print("Result =", conversions[choice](val))
        else:
            print("Invalid choice!")

    # ------------------------------ PART 9: INVALID OPTION HANDLING ------------------------------ #
    else:
        # Handle invalid main menu input
        print("Invalid option!")

# ------------------------------ PART 10: FUNCTION CALL ------------------------------ #
# Call the function to run the program
measurements()
```

**Explanation**

- I created a **function** called **`measurements()`** to hold all the unit conversion logic, making the program **reusable**, **organized**, and **easy to maintain**.
   
- Inside the function, a **main menu** is printed using `print()` showing six categories: **Length, Temperature, Mass, Volume, Pressure, and Energy**.  
- The user enters a choice using **`input()`**, stored in the variable **`option`**, which determines which category the program will process.

- The program uses **`if-elif-else` statements** to check the user input and execute the correct category. If the input is not between 1–6, it prints **"Invalid option!"**.

- Each category displays a **sub-menu** with specific conversion options. Example for Length: **Meters → Feet**, **Inches → Centimeters**, **Miles → Kilometers**.

- A **dictionary** named **`conversions`** stores the formulas for each sub-menu choice, with **lambda functions** performing the calculation. Example: `"1": lambda x: x * 3.28084`.

- Using **lambda functions** makes the code **compact**, avoids multiple function definitions, and allows **quick calculation** inline.

- The user inputs the numeric value for conversion using **`float()`**, which allows **decimal numbers**.

- The program checks if the user’s **sub-menu choice exists** in the `conversions` dictionary. If valid, the **lambda function** is called with the input, and the result is printed. Otherwise, it prints **"Invalid choice!"**.

- The same structure is repeated for other categories: **Temperature:** Celsius → Fahrenheit, Celsius → Kelvin; **Mass:** Kilograms → Pounds, Grams → Ounces; **Volume:** Liters → Gallons, Milliliters → Ounces; **Pressure:** Pascal → Bar, PSI → kPa; **Energy:** Joules → Calories, Kilojoules → Kilocalories.

- The program includes **error handling** for both invalid main menu options and invalid sub-menu choices, making it **robust** and **user-friendly**.  
- Finally, the function **`measurements()`** is called at the end to run the program, demonstrating how **functions can organize and execute reusable code** efficiently.

- This lab showcases **functions**, **lambda expressions**, **dictionaries for mapping choices to actions**, **input validation**, and **menu-driven programming**, all combining to create an **efficient and reusable program** in Python.

```python
import re   # Importing the regular expression module

while True:
    # Ask user to enter a password
    password = input("Enter your password: ")

    # Check if password contains spaces
    if " " in password:
        print("Password should not contain spaces.")

    # Check if password length is more than 8 characters
    elif len(password) <= 8:
        print("Password must be more than 8 characters.")

    # Check if password contains at least one uppercase letter
    elif not re.search(r"[A-Z]", password):
        print("Password must have at least one uppercase letter.")

    # Check if password contains at least one lowercase letter
    elif not re.search(r"[a-z]", password):
        print("Password must have at least one lowercase letter.")

    # Check if password contains at least one digit
    elif not re.search(r"\d", password):
        print("Password must have at least one number.")

    # Check if password contains at least one special character
    elif not re.search(r"[@#$%^&*!?]", password):
        print("Password must have at least one special character (@, #, $, etc.).")

    # Passed all validations
    else:
        print("Password is valid.")
        break   # Exit the loop
```

**Explanation**

- I started by importing the **re module**, which allows the program to use **regular expressions** for checking patterns in text.

- Then, I used a **while True loop** so the program keeps asking for a password until the user enters a valid one.

- Inside the loop, I used **input()** to ask the user to enter a password and stored it in a variable called **password**.

- I first checked if there is a **space (" ") inside the password**. If yes, the program tells the user that spaces are not allowed.

- Next, I used **len(password)** to make sure the password is **more than 8 characters long**. If it is not long enough, an error message is shown.

- After that, I used **re.search(r"[A-Z]", password)** to check if the password has at least **one uppercase letter**.

- Then, I used **re.search(r"[a-z]", password)** to check if the password contains at least **one lowercase letter**.

- To check if the password includes a **digit**, I used **re.search(r"\d", password)**.

- I also checked for at least **one special character** such as `@, #, $, %, ^, &, *, !, ?` using **re.search(r"[@#$%^&*!?]", password)**.

- All these conditions were checked using **if-elif statements**, which make the program run each check one by one and show an error message if something is missing.

- If any rule is broken, the program shows the message telling the user what is wrong, and the loop continues.

- If all conditions are met, the program prints **"Password is valid."**

- Finally, I used **break** to exit the loop once a valid password is entered.

```python
# ------------------------------ Create Initial Clothing Stock ------------------------------ #
clothing_stock = """Hoodie, XS
Hoodie, S
Hoodie, M
Hoodie, L
Hoodie, XL
Hoodie, XXL
Hoodie, XL
Hoodie, XXL
Suit, XS
Suit, S
Suit, M
Suit, L
Suit, XL
Suit, L
Suit, XL
Suit, XXL
Jeans, XS
Jeans, S
Jeans, M
Jeans, L
Jeans, XL
Jeans, XXL
Shirt, XS
Shirt, S
Shirt, M
Shirt, L
Shirt, XL
Shirt, XXL
T-shirt, XS
T-shirt, S
T-shirt, M
T-shirt, L
T-shirt, XL
T-shirt, XXL
Legging, XS
Legging, S
Legging, M
Legging, L
Legging, XL
Legging, XXL
Jacket, XS
Jacket, S
Jacket, M
Jacket, L
Jacket, XL
Jacket, XL
Jacket, XXL
Trousers, XS
Trousers, S
Trousers, M
Trousers, L
Trousers, XL
Trousers, XXL"""

# Write the initial stock to 'clothing_stock.txt'
with open("clothing_stock.txt", "w") as file:
    file.write(clothing_stock)

print("Initial clothing stock has been created in 'clothing_stock.txt'.")

# ------------------------------ Read Clothing Stock ------------------------------ #
try:
    with open("clothing_stock.txt", "r") as file:
        # Read all lines and remove newline characters
        clothing_list = [line.strip() for line in file]
except FileNotFoundError:
    print("Error: 'clothing_stock.txt' not found.")
    clothing_list = []

# ------------------------------ Remove Duplicates ------------------------------ #
# Use a set to remove duplicates
unique_clothing = list(set(clothing_list))

# Optional: sort the list alphabetically
unique_clothing.sort()

# ------------------------------ Overwrite File with Unique Entries ------------------------------ #
with open("clothing_stock.txt", "w") as file:
    for item in unique_clothing:
        file.write(item + "\n")

print("Duplicate items removed. 'clothing_stock.txt' now contains only unique entries.")
```

**Explanation**

- First, I created a **multi-line string** called **clothing_stock** that contains all clothing items with their sizes, including duplicates.  
  This simulates the initial stock file that we want to clean.

- Next, I used a **`with open()` statement in write mode (`"w"`)** to create a file named `clothing_stock.txt` and write the contents of `clothing_stock` to it.  
  - This step demonstrates **file handling** in Python.  
  - Using `with` ensures that the file is properly closed after writing, which is good practice.

- I printed a message to confirm that the initial file has been created.

- After creating the file, I used a **try-except block** to read from it.  
  - `try:` attempts to open the file in read mode (`"r"`) and read all lines.  
  - Each line is processed with **`strip()`** to remove newline characters and stored in a **list** called `clothing_list`.  
  - If the file does not exist, the **`except FileNotFoundError`** block prints an error message and sets `clothing_list` to an empty list.  
  - This shows how **exceptions** can be used to make the program robust and avoid crashes.

- To remove duplicates, I converted the list into a **set** using `set(clothing_list)`.  
  - Sets are Python **data structures** that store only unique items.  
  - By converting the list to a set, all duplicate clothing items (e.g., multiple entries of "Hoodie, XL") are automatically removed.

- I then converted the set back to a **list** to allow further manipulation, such as sorting.

- I optionally sorted the list alphabetically using `unique_clothing.sort()`.  
  - Sorting is useful for organizing the inventory, making it easier to read and manage.

- Finally, I opened the same file `"clothing_stock.txt"` in **write mode (`"w"`)** again to overwrite it with only unique entries.  
  - I looped through each item in `unique_clothing` and wrote it to the file, adding a newline character after each item.  
  - This step demonstrates **both file handling and data processing** by saving the cleaned, unique stock back into the original file.

- I printed a confirmation message to indicate that duplicates were removed and the file now contains only unique items.
